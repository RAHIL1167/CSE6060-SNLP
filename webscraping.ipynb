```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq

my_url = 'http://www.imdb.com/search/title?release_date=2017&sort=num_votes,desc&page=1'

uClient = uReq(my_url)
page_html=uClient.read()
uClient.close()
page_soup= soup(page_html,"html.parser")
```

```python
movie_containers=page_soup.findAll("div",class_='lister-item mode-advanced')
```

```python
print(len(movie_containers)) #no of pages displayed on that page
```

Now we’ll select only the first container, and extract, by turn, each item of interest:

The name of the movie.

The year of release.

The IMDB rating.

The Metascore.

The number of votes.

```python
# Lists to store the scraped data in
names = []
years = []
imdb_ratings = []
metascores = []
votes = []
# Extract data from individual movie container
for container in movie_containers:
# If the movie has Metascore, then extract:
    if container.find('div', class_ = 'ratings-metascore') is not None:
# The name
        name = container.h3.a.text
        names.append(name)
# The year
        year = container.h3.find('span', class_ = 'lister-item-year').text
        years.append(year)
# The IMDB rating
        imdb = float(container.strong.text)
        imdb_ratings.append(imdb)
# The Metascore
        m_score = container.find('span', class_ = 'metascore').text
        metascores.append(int(m_score))
# The number of votes
        vote = container.find('span', attrs = {'name':'nv'})['data-value']
        votes.append(int(vote))
```

```python
import pandas as pd
test_df = pd.DataFrame({'movie': names,
'year': years,
'imdb': imdb_ratings,
'metascore': metascores,
'votes': votes
})
print(test_df.info())
test_df
```

```python
movie_ratings = pd.DataFrame({'movie': names,
'year': years,
'imdb': imdb_ratings,
'metascore': metascores,
'votes': votes
})
print(movie_ratings.info())
movie_ratings.head(10)
```

# Extracting named entity from an article


```python
from bs4 import BeautifulSoup
import requests
import re
def url_to_string(url):
    res = requests.get(url)
    html = res.text
    soup = BeautifulSoup(html, 'html5lib')
    for script in soup(["script", "style", 'aside']):
        script.extract()
    return " ".join(re.split(r'[\n\t]+', soup.get_text()))
ny_bb = url_to_string('https://towardsdatascience.com/nlp-for-indian-languages-310d1d8a10b6')
article = nlp(ny_bb)
len(article.ents)
```

There are 251 entities present in the article

```python
labels = [x.label_ for x in article.ents]

Counter(labels)
```

```python
#Most 3 frequent tokens

items = [x.text for x in article.ents]
Counter(items).most_common(3)
```

```python
sentences = [x for x in article.sents]
print(sentences[20])
```

```python
displacy.render(nlp(str(sentences[20])), jupyter=True, style='ent')
```

```python
displacy.render(nlp(str(sentences[20])), style='dep', jupyter = True, options = {'distance': 120})
```
