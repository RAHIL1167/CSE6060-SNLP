```python
import nltk
import docx2txt
```

```python
lor1 =  docx2txt.process("LoR1.docx")
lor2 =  docx2txt.process("LoR2.docx")
```

```python
print(lor1)
```

```python
grammar = "NP: {<DT>?<JJ>*<NN>}"
grammar1 = "VP: {<PRP>+<VB>+<JJ>?}"
```

```python
cp = nltk.RegexpParser(grammar)
cp1 = nltk.RegexpParser(grammar1)
```

```python
tokens = nltk.word_tokenize(lor1)
print(tokens)
```

```python
pos_tags = nltk.pos_tag(tokens)
print(pos_tags)
```

```python
chunks_lor1 = cp.parse(pos_tags)
print(chunks_lor1)
```

```python
chunks_lor1.draw()
```

```python
chunks_lor12 = cp1.parse(pos_tags)
print(chunks_lor12)
```

```python
chunks_lor12.draw()
```

```python
chunks = nltk.ne_chunk(pos_tags)
```

```python
chunks.draw()
```

```python
example = "Google, a company founded by Larry Page and Sergey Brin in the United States of America "\
+ "has one of the world’s most advanced search engines."
doc=nlp(example)

for ent in doc.ents:
    print(ent.text, ent.label_)
```

```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq

my_url = 'http://www.imdb.com/search/title?release_date=2017&sort=num_votes,desc&page=1'

uClient = uReq(my_url)
page_html=uClient.read()
uClient.close()
page_soup= soup(page_html,"html.parser")
```

```python
containers=page_soup.findAll("div",class_='lister-item mode-advanced')
```

```python
print(len(containers)) #no of pages displayed on that page
```

```python
print(soup.prettify(containers[0]))
```
