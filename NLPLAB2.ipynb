# Task2 Stemming:


In Stemming we reduce the word to it's original form by removing suffixes and prefixes in the word for example code, coder, coders, conding. all these belong to the same word code

we may have errors in stemming like over stemming and under stemming



In overstemming too much of word is cut and meaning is lost. In this words of 2 different stems are reduced to same stem such as university, universal they are reduced univers infact they belong to different stems

In understemming two words of same stem are mapped to different stem
such as data is reduced dat and datum is reduced to datu

Porterstemmer is rule based stemmer

# Task3: Stemming Paragraphs


```python
#HW Demonstrate snowball stemming in hindi/tamil/french
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
example= "Am quick brown fox jumps over a lazy dog"
example=[stemmer.stem(token) for token in example.split(" ")]
print(" ".join(example))
```

# task4 :Lemmatizer


Lemmatization is similar to stemming but is accurate and slower than stemming.


```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
print(lemmatizer.lemmatize("cacti"))
```

wordnet is a collection of verbs, adjectivec, adverbs grouped together with all synonyms of the word

```python
print(lemmatizer.lemmatize("potterer"))
```

```python
print(lemmatizer.lemmatize("better",pos='a'))
```

```python
print(lemmatizer.lemmatize("am",pos='v'))
```

# Task 6: BASIC TEXT PROCESSING PIPELINE


```python
import nltk
sent=" Become an expert in NLP"
words=nltk.word_tokenize(sent)
print(words)
```

```python
texts=""""PlayerUnknown" Greene for other games, inspired by the 2000 Japanese film Battle Royale, and expanded into a standalone game under Greene's creative direction. In the game, up to one hundred players parachute onto an island and scavenge for weapons and equipment to kill others while avoiding getting killed themselves. The available safe area of the game's map decreases in size over time, directing surviving players into tighter areas to force encounters. The last player or team standing wins the round."""
```

```python
for text in texts:
    sentences=nltk.sent_tokenize(texts)
    for sentence in sentences:
        words=nltk.word_tokenize(sentence)
        print(words)
        #tagged=nltk.pos_tag(words)
        #print(tagged)
```

```python
tagged=nltk.pos_tag(words)
print(tagged)
```
