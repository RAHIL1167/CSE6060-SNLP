```python
text1='''Vice President Johnson , Mr . Speaker , Mr . Chief Justice , President Eisenhower , Vice President Nixon , President Truman , reverend clergy , fellow citizens , we observe today not a victory of party , but a celebration of freedom -- symbolizing an end , as well as a beginning -- signifying renewal , as well as change . For I have sworn I before you and Almighty God the same solemn oath our forebears l prescribed nearly a century and three quarters ago . The world is very different now .'''
fd=nltk.FreqDist(text1.split())
fd
```

```python
from nltk.probability import ConditionalFreqDist
cfd=ConditionalFreqDist((len(word),word) for word in text1.split())

        
```

Finding  conditional frequency for 10 letter, 8 letter, 6 letter words 

```python
cfd[8]
```

```python
cfd[6]
```

```python
cfd[10]
```

```python
sort_fd = sorted(fd.items(), key = lambda x:x[1])
fw = list(sort_fd)[len(sort_fd)-1]
print("Frequently used word :",fw)
```
